```python
# Create Column For ActualAmount And ActualSales
data['ActualAmount']=(data['QUANTITYORDERED'])*(data['ManufPrice'])
data['ActualSales']=(data['QUANTITYORDERED'])*(data['PRICEEACH'])
print(data.head())
```

```python
#Difference ActualAmount & ActualSales
data['ProfitLoss'] = (data['ActualSales'])-(data['ActualAmount'])
print(data['ProfitLoss'])
```

```python
#Create new column if value is negative then Loss and if value is positive then profit signword
data['signword'] = ['profit' if i>=0 else 'loss' for i in data['ProfitLoss'] ]
data['bincode'] = [1 if i>=0 else 0 for i in data['ProfitLoss']]
print(data['signword'],data['bincode'])
```

```python
data.to_csv("AutoSales_1.csv",index=False)
```

```python
data.head()
```

```python
#Calculate the Difference for loss or profit
AmountSum = data['ActualAmount'].sum()
SalesSum = data['ActualSales'].sum()
print(f"The Sum of TotalAmount is {AmountSum}")
print(f"The Sum of TotalSales is {SalesSum}")
# All over calculate loss or profit (- sign->Loss & + sign->Profit)
print(SalesSum-AmountSum)
print(data['ProfitLoss'].sum())
```
